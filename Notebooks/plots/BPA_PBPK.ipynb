```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load the simulation results
with open('result.json', 'r') as f:
    data = json.load(f)

# Extract data
time = np.array(data['time'])
aplasma = np.array(data['species']['aplasma'])

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(time, aplasma, 'b-', linewidth=2, label='Aplasma')
plt.xlabel('Time (hours)', fontsize=12)
plt.ylabel('Concentration', fontsize=12)
plt.title('Aplasma Concentration vs Time', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
